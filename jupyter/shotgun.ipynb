```python
import numpy as np

def get_tree_vertices(tree_depth = 3):
    # Each kin has 2 kins
    # Each kin has 1 parent
    tree_vertices = 0
    for i in range(tree_depth):
        # Add 2^i vertices
        tree_vertices += 2**i
    return tree_vertices

def create_tree(tree_depth=3):
    assert tree_depth > 1, "Minimum depth of the tree is 2"
    
    tree_vertices = get_tree_vertices(tree_depth)
    # print(f"Tree has {tree_vertices} vertices")
    A = np.zeros((tree_vertices,tree_vertices), dtype=float) # Adjacency matrix
    
    vertices_added = []
    # print(f"{tree_vertices-2**(tree_depth-1)}")
    for i in range(tree_vertices-2**(tree_depth-1)):
        A[i,i] = 1
        vertices_added.append(i)
        # Add 2 new ones
        new_vertix = max(vertices_added)+1
        # print(f"New kins can be {new_vertix} and {new_vertix+1}")
        A[i,new_vertix] = 1
        A[new_vertix,i] = 1
        A[new_vertix,new_vertix] = 1
        
        A[i,new_vertix+1] = 1
        A[new_vertix+1,i] = 1
        A[new_vertix+1,new_vertix+1] = 1
        
        vertices_added.append(new_vertix)
        vertices_added.append(new_vertix+1)

    return A

def create_p_weights(A, p=0.4):
    P = np.full_like(A, 0) # Edge probabilites
    I,J = P.shape
    for i in range(I):
        for j in range(J):
            if i == j:
                continue # Don't weigh self
            if i > j:
                continue # Don't weigh parent connection
            if A[i,j] == 1:
                # print(f"{i},{j} is a kin connection")
                if j % 2: # the way tree is created, one kin is always even, one is odd
                    P[i,j] = p
                    P[j,i] = p
                else:
                    P[i,j] = 1-p
                    P[j,i] = 1-p
    return P

def get_parent(v,A):
    I,J = A.shape
    for i in range(I+1):
        if A[v,i] == 1:
            return i

def get_p_edge_line(A,P, tree_depth = 3, p=0.4):
    # Edge line means a "line" created by traversing edges from the last kin through its parents to the root
    
    # Last kins have indices
    tree_vertices,_ = A.shape
    last_kins_num = 2**(tree_depth-1)
    last_kins_first_id = tree_vertices - last_kins_num
    # print(f"last_kins_first_id is {last_kins_first_id}")
    
    edge_lines = []
    for i in range(last_kins_first_id,tree_vertices):
        kin = i
        partial_p = 1
        num_success = 0
        while (kin): # root is 0
            parent = get_parent(kin,A)
            # print(f"Found parent {parent} of kin {kin}")
            # print(f"This connection has weight {P[parent,kin]}")
            partial_p = partial_p*P[parent,kin]
            # print(f"New kin will be {parent}")
            if kin%2:
                num_success += 1
            kin = parent
        if partial_p == 1:
            print(f"WARNING: partial_p is equal to 1!")
        edge_lines.append([partial_p, num_success])
    return edge_lines

def print_stats(edge_lines):
    v = []
    for item in edge_lines:
        v.append(item[1])
    max_success = max(v)
    p_success_i = []
    for i in range(max_success+1):
        p_success = 0
        for item in edge_lines:
            if item[1] == i:
                p_success += item[0]
        p_success_i.append([100*p_success,i])
    print("Number of successes, Chance [%] ")
    for i in p_success_i:
        print(f"{i[1]} {i[0]:8.3f}")


guard_skill_sweep = [0.1, 0.15, 0.2, 0.25, 0.4, 0.65]
num_shots_sweep = [1,3,5,9,10]
for guard_skill in guard_skill_sweep:
    p = guard_skill
    for num_shots in num_shots_sweep:
        print("------"*10)
        print(f"Number of taken shots: {num_shots}")
        tree_depth = num_shots+1
        print(f"Guard skill: {p}")
        A = create_tree(tree_depth)
        # print(A)
        P = create_p_weights(A, p)
        edge_lines = get_p_edge_line(A,P,tree_depth,p)
        # print(f"edge_lines is {edge_lines}")
        print_stats(edge_lines)
```
